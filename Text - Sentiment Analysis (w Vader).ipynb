### Sentiment using NLTK Vader

 Vader is lexicon and rule-based sentiment analysis package that is specifically attuned to sentiments expressed in social media

```python
import pandas as pd
```

```python
# read file
df = pd.read_csv("fashion.csv",header = 0)

df.head()
```

```python
#import nltk vader
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer #Vader 

#set list for each polarity
dictn = []
dictu = []
dictp = []
#dictcom = []

#save the value of quality measurements
sid = SentimentIntensityAnalyzer()
for sentence in df['review text']:
    ss = sid.polarity_scores(sentence)
    for k in ss:
            if k == 'neg':
                t=ss[k]
                dictn.append(t)
            elif k == 'neu':
                t=ss[k]
                dictu.append(t)
            elif k == 'pos':
                t=ss[k]
                dictp.append(t)
            #elif k == 'compound':
                #t=ss[k]
                #dictcom.append(t)
            print('{0}: {1},  '.format(k, ss[k]), end='')
    print()
    
```

```python
from pandas import DataFrame
#combine each score build a new data frame
dictc=DataFrame()

dictc['ID'] = list(range(1,len(df)+1))
df['ID'] = list(range(1,len(df)+1))
dictc['positive'] = dictp
dictc['negative'] = dictn
dictc['neutral'] = dictu
#dictc['compound'] = dictcom
dictc
```

```python
#write result into csv
dictc.to_csv('sentiment_output.csv')
```

```python
#merge two datasets
merge_dataset = pd.merge(df,dictc,on="ID")
merge_dataset
```
